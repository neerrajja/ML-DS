```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
mydata=pd.read_csv('Mall_Customers.csv')

X=mydata.iloc[:,[3,4]].values

from sklearn.cluster import KMeans
km=KMeans(n_clusters=4)
km.fit(X)
pred=km.predict(X)
print(pred)

plt.scatter(X[pred==0,0],X[pred==0,1],c='blue',label='customer 1')
plt.scatter(X[pred==1,0],X[pred==1,1],c='blue',label='customer 2')
plt.scatter(X[pred==2,0],X[pred==2,1],c='blue',label='customer 3')
plt.scatter(X[pred==3,0],X[pred==3,1],c='blue',label='customer 4')

plt.legend()
plt.show()

plt.scatter(X[:,0],X[:,1],c=pred)
plt.show()
```
